# Concept Similarity

```python
import warnings
warnings.filterwarnings('ignore')
warnings.simplefilter('ignore')
from torchvision.models.segmentation import deeplabv3_resnet50
import torch
import torch.functional as F
import numpy as np
import requests
import cv2
import torchvision
from PIL import Image
from pytorch_grad_cam.utils.image import show_cam_on_image, preprocess_image
from pytorch_grad_cam import GradCAM
from utils.image_loader import get_image_from_fs


# A model wrapper that gets a resnet model and returns the features before the fully connected layer.
class ResnetFeatureExtractor(torch.nn.Module):
    def __init__(self, model):
        super(ResnetFeatureExtractor, self).__init__()
        self.model = model
        self.feature_extractor = torch.nn.Sequential(*list(self.model.children())[:-1])
                
    def __call__(self, x):
        return self.feature_extractor(x)[:, :, 0, 0]
    
class SimilarityToConceptTarget:
    def __init__(self, features):
        self.features = features
    
    def __call__(self, model_output):
        cos = torch.nn.CosineSimilarity(dim=0)
        return cos(model_output, self.features)
        
resnet = torchvision.models.resnet50(pretrained=True)
resnet.eval()
model = ResnetFeatureExtractor(resnet)
```

```python
dir_name = "images/website"
reference_img_name = "screenshot4.jpg"
concept_img_name = "icon.jpg"

reference_img, reference_img_float, reference_img_tensor = get_image_from_fs(f"{dir_name}/{reference_img_name}")
reference_img_features = model(reference_img_tensor)[0, :]

concept_img, concept_img_float, concept_img_tensor = get_image_from_fs(f"{dir_name}/concepts/{concept_img_name}")
concept_img_features = model(concept_img_tensor)[0, :]
```

```python
Image.fromarray(reference_img)
```

```python
target_layers = [resnet.layer4[-1]]
reference_img_targets = [SimilarityToConceptTarget(concept_img_features)]

# Where is the car in the image
with GradCAM(model=model,
             target_layers=target_layers,
             use_cuda=False) as cam:
    grayscale_cam = cam(input_tensor=reference_img_tensor,
                        targets=reference_img_targets)[0, :]
cam_image = show_cam_on_image(reference_img_float, grayscale_cam, use_rgb=True)
Image.fromarray(cam_image)
```
